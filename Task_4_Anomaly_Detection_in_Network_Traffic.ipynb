```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
for column in ['Protocol', 'SourceIP', 'DestinationIP']:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
df
```

```python
features = ['Duration', 'Protocol', 'SourcePort', 'DestinationPort', 'PacketCount', 'ByteCount']
final_data = df[features]
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(final_data)
```

```python
df
```

```python
from sklearn.ensemble import IsolationForest
model = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)
model.fit(scaled_data)
df['anomaly'] = model.predict(scaled_data)
df['anomaly'] = df['anomaly'].map({1: 0, -1: 1})
```

```python
anomalies = df[df['anomaly'] == 1]
normal = df[df['anomaly'] == 0]
# Visualize anomalies in a scatter plot of two selected features (e.g., Duration vs PacketCount)
plt.figure(figsize=(10, 6))
plt.scatter(normal['Duration'], normal['PacketCount'], c='blue', label='Normal')
plt.scatter(anomalies['Duration'], anomalies['PacketCount'], c='red', label='Anomaly')
plt.xlabel('Duration')
plt.ylabel('Packet Count')
plt.legend()
plt.title('Anomalies in Network Traffic')
plt.show()
```

```python
df['PacketSizePerDuration'] = df['ByteSizePerPacket'] / (df['PacketSizePerDuration']+ 1e-9)
df['ByteSizePerPacket'] = df['ByteSizePerPacket'] / (df['PacketSizePerDuration']+ 1e-9)

anomalies = df[df['anomaly'] == 1]
normal = df[df['anomaly'] == 0]
plt.figure(figsize=(10, 6))
plt.scatter(normal['PacketSizePerDuration'], normal['ByteSizePerPacket'], c='green', label='Normal')
plt.scatter(anomalies['PacketSizePerDuration'], anomalies['ByteSizePerPacket'], c='yellow', label='Anomaly')
plt.xlabel('PacketSizePerDuration')
plt.ylabel('ByteSizePerPacket')
plt.legend()
plt.title('Anomalies in Network Traffic')
plt.show()
```
